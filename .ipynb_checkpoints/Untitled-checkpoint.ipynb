```python
from PIL import Image
import matplotlib.pyplot as plt
import torchvision.transforms as transforms
import torchvision.transforms.functional as F

# Path to the example slice
path = "/home/jovyan/Desktop/CGMH-GBM-2D-slice/GP002/t1/sagittal/slice_0085.png"

# 1) Load original image
img = Image.open(path).convert("RGB")

# 2) Resize long side to 224 then center crop 224x224
transform_resize_crop = transforms.Compose([
    transforms.Resize(224)]
    
                                          
)

img_rc = transform_resize_crop(img)

# 3) Pad to square then resize to 224x224
def pad_to_square(pil_img):
    w, h = pil_img.size
    diff = abs(h - w)
    pad_left = pad_top = 0
    pad_right = pad_bottom = 0
    if h < w:
        # pad height
        pad_top = diff // 2
        pad_bottom = diff - pad_top
    else:
        # pad width
        pad_left = diff // 2
        pad_right = diff - pad_left
    return F.pad(pil_img, (pad_left, pad_top, pad_right, pad_bottom), fill=0)

img_pad = pad_to_square(img)
img_pad_resized = transforms.Resize((224, 224))(img_pad)

# 4) Plot all images side by side
fig, axes = plt.subplots(1, 4, figsize=(16, 4))

axes[0].imshow(img)
axes[0].set_title(f"Original\n{img.size[0]}×{img.size[1]}")
axes[0].axis('off')

axes[1].imshow(img_rc)
axes[1].set_title("Resize→CenterCrop\n224×224")
axes[1].axis('off')

axes[2].imshow(img_pad)
axes[2].set_title(f"Padded Square\n{img_pad.size[0]}×{img_pad.size[1]}")
axes[2].axis('off')

axes[3].imshow(img_pad_resized)
axes[3].set_title("Pad→Resize\n224×224")
axes[3].axis('off')

plt.tight_layout()
plt.show()
```
